### Hypothesis Testing

##### H: Promotion tidak berpengaruh terhadap Units Sold
##### H: Promotion berpengaruh terhadap Units Sold

#### Pengaruh Promosi terhadap Penjualan

```python
from scipy.stats import ttest_ind

promo = df[df["Promotion"] == 1]["Units Sold"]
non_promo = df[df["Promotion"] == 0]["Units Sold"]

t_stat, p_value = ttest_ind(promo, non_promo, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

#### Dampak Epidemi terhadap Penjualan

##### H: Epidemi tidak mempengaruhi Units Sold
##### H: Epidemi mempengaruhi Units Sold

```python
epidemic = df[df["Epidemic"] == 1]["Units Sold"]
normal = df[df["Epidemic"] == 0]["Units Sold"]

t_stat, p_value = ttest_ind(epidemic, normal, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

#### Perbedaan Penjualan Antar Kategori Produk

##### H: Tidak ada perbedaan Units Sold antar kategori

##### H: Ada perbedaan Units Sold antar kategori

```python
from scipy.stats import f_oneway

groups = [
    df[df["Category"] == cat]["Units Sold"]
    for cat in df["Category"].unique()
]

f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("p-value:", p_value)
```

### Feature Engineering & Preprocessing

#### Seleksi Fitur (BERDASARKAN EDA & UJI HIPOTESIS)

```python
numeric_features = [
    "Inventory Level",
    "Units Ordered",
    "Price",
    "Discount",
    "Competitor Pricing",
    "Demand"
]
```

```python
binary_features = [
    "Promotion",
    "Epidemic"
]
```

```python
categorical_features = [
    "Category",
    "Region",
    "Weather Condition",
    "Seasonality"
]
```

#### Encoding Kategori

```python
from sklearn.preprocessing import LabelEncoder

df_fe = df.copy()

for col in categorical_features:
    le = LabelEncoder()
    df_fe[col] = le.fit_transform(df_fe[col])
```

```python
df_fe.isna().sum()
```

#### Aggregasi Harian

```python
features = numeric_features + binary_features + categorical_features

daily_df = (
    df_fe
    .groupby("Date")[["Units Sold"] + features]
    .mean()
    .sort_index()
)

daily_df.head()
```

#### TrainTest Split (Time-Based)

```python
train_size = int(len(daily_df) * 0.8)

train_df = daily_df.iloc[:train_size]
test_df  = daily_df.iloc[train_size:]
```

```python
print("Train:", train_df.shape)
print("Test :", test_df.shape)
```

#### Scaling (NO DATA LEAKAGE)

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_train = scaler.fit_transform(train_df)
scaled_test  = scaler.transform(test_df)
```

### MODELING LSTM

#### Time Series Generator

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

n_input = 14          
batch_size = 32

n_features = scaled_train.shape[1]

train_generator = TimeseriesGenerator(
    scaled_train,
    scaled_train[:, 0], #unit sold
    length=n_input,
    batch_size=batch_size
)
```

```python
len(train_generator)
```

#### Arsitektur Model LSTM

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(n_input, n_features)),
    Dropout(0.2),
    LSTM(32),
    Dense(1)
])

model.compile(
    optimizer="adam",
    loss="mse"
)

model.summary()
```

```python
history = model.fit(
    train_generator,
    epochs=30,
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history["loss"])
plt.title("Training Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

#### Forecasting (Predict Data Test)

```python
import numpy as np

predictions = []

last_train_batch = scaled_train[-n_input:]
current_batch = last_train_batch.reshape((1, n_input, n_features))

for i in range(len(test_df)):
    pred = model.predict(current_batch, verbose=0)[0]
    predictions.append(pred)

    current_batch = np.append(
        current_batch[:, 1:, :],
        [[np.concatenate([pred, scaled_test[i, 1:]])]],
        axis=1
    )
```

#### Inverse Scaling (Kembali ke Skala Asli)

```python
pred_array = np.zeros((len(predictions), n_features))
pred_array[:, 0] = np.array(predictions).flatten()

true_predictions = scaler.inverse_transform(pred_array)[:, 0]
```
