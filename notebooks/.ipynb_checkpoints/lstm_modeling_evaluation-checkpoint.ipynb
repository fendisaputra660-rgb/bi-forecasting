```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import joblib
```

```python
daily_df = pd.read_csv("../data/processed/daily_features.csv", index_col=0, parse_dates=True)
```

```python
daily_df.head()
```

```python
daily_df.info()
```

```python
train_size = int(len(daily_df) * 0.8)

train_df = daily_df.iloc[:train_size]
test_df  = daily_df.iloc[train_size:]

print(train_df.shape, test_df.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_train = scaler.fit_transform(train_df)
scaled_test  = scaler.transform(test_df)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

n_input = 14
n_features = scaled_train.shape[1]

train_generator = TimeseriesGenerator(
    scaled_train,
    scaled_train[:, 0],   # target = Units Sold
    length=n_input,
    batch_size=32
)

len(train_generator)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(n_input, n_features)),
    Dropout(0.2),
    LSTM(32),
    Dense(1)
])

model.compile(
    optimizer="adam",
    loss="mse"
)

model.summary()
```

```python
history = model.fit(
    train_generator,
    epochs=30,
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6)) 
plt.plot(history.history["loss"])
plt.title("Training Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
import numpy as np

predictions = []

last_batch = scaled_train[-n_input:]
current_batch = last_batch.reshape((1, n_input, n_features))

for i in range(len(test_df)):
    pred = model.predict(current_batch, verbose=0)[0]
    predictions.append(pred)

    current_batch = np.append(
        current_batch[:, 1:, :],
        [[np.concatenate([pred, scaled_test[i, 1:]])]],
        axis=1
    )
```

```python
pred_array = np.zeros((len(predictions), n_features))
pred_array[:, 0] = np.array(predictions).flatten()

true_predictions = scaler.inverse_transform(pred_array)[:, 0]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

actual = test_df.iloc[:, 0].values

rmse = np.sqrt(mean_squared_error(actual, true_predictions))
mae  = mean_absolute_error(actual, true_predictions)

print("RMSE:", rmse)
print("MAE :", mae)
print("Mean Units Sold:", actual.mean())
```

```python
plt.figure(figsize=(12,6))
plt.plot(test_df.index, actual, label="Actual")
plt.plot(test_df.index, true_predictions, label="Prediction")
plt.title("Actual vs Predicted Units Sold")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.legend()
plt.show()
```
