```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/sales_data.csv")
df.head()
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

DATE_COL = "Date"
TARGET_COL = "Units Sold"

print(df[DATE_COL].min(), df[DATE_COL].max())
```

```python
plt.figure()
df[TARGET_COL].hist(bins=40)
plt.title("Distribution of Units Sold")
plt.show()
```

```python
daily_units = df.groupby(DATE_COL)[TARGET_COL].sum()

plt.figure(figsize=(12,4))
daily_units.plot()
plt.title("Daily Units Sold Trend")
plt.show()
```

```python
weekly_units = daily_units.resample("W").sum()
monthly_units = daily_units.resample("M").sum()

plt.figure(figsize=(12,4))
weekly_units.plot(title="Weekly Units Sold")
plt.show()

plt.figure(figsize=(12,4))
monthly_units.plot(title="Monthly Units Sold")
plt.show()
```

```python
df.groupby("Category")[TARGET_COL].mean().sort_values(ascending=False)
```

```python
df.groupby("Region")[TARGET_COL].mean().sort_values(ascending=False)
```

```python
df.groupby("Promotion")[TARGET_COL].mean()
```

```python
df.groupby("Weather Condition")[TARGET_COL].mean()
```

## EDA 

- Penjualan menunjukkan pola musiman pada level mingguan dan bulanan.
- Terdapat perbedaan penjualan antar kategori dan wilayah.
- Promosi dan faktor eksternal berasosiasi dengan peningkatan Units Sold.
