# Bokeh & Flask Integration Tutorial
## Cara Membuat Dashboard Interaktif dari Jupyter ke Flask Web App

Notebook ini mendemonstrasikan bagaimana mengexport visualisasi Bokeh dari Jupyter notebook ke Flask application.

## 1. Import Required Libraries

Impor library yang diperlukan untuk membuat visualisasi interaktif dengan Bokeh.

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import os
import sys

# Bokeh imports untuk visualisasi interaktif
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import HoverTool, ColumnDataSource, Slider, Select
from bokeh.layouts import column, row, gridplot
from bokeh.palettes import Category20
from bokeh.transform import cumsum
from bokeh.io import export_png, export_svgs
from bokeh.embed import components
from bokeh.models.callbacks import CustomJS

# Data processing
import matplotlib.pyplot as plt

print("âœ“ Semua library berhasil diimport")
```

## 2. Load Data dari CSV

Membaca data dari file CSV yang juga digunakan oleh Flask app.

```python
# Tentukan path ke data
BASE_DIR = os.path.abspath("../data/processed")
csv_file = os.path.join(BASE_DIR, "daily_features.csv")

# Load data
if os.path.exists(csv_file):
    df = pd.read_csv(csv_file, parse_dates=["Date"])
    print(f"âœ“ Data berhasil dimuat: {df.shape[0]} baris, {df.shape[1]} kolom")
    print(f"\nKolom yang tersedia:\n{df.columns.tolist()}")
    print(f"\nSample data:\n{df.head()}")
else:
    print(f"âŒ File tidak ditemukan: {csv_file}")
    print("Menggunakan data dummy untuk demonstrasi...")
    # Buat data dummy
    dates = pd.date_range(start='2024-01-01', periods=100, freq='D')
    df = pd.DataFrame({
        'Date': dates,
        'Units Sold': np.random.randint(50, 200, 100),
        'Price': np.random.uniform(10, 50, 100),
        'Category': np.random.choice(['A', 'B', 'C'], 100),
        'Region': np.random.choice(['North', 'South', 'East', 'West'], 100),
        'Promotion': np.random.choice([0, 1], 100)
    })
```

## 3. Configure Bokeh Output in Jupyter

Konfigurasi Bokeh untuk menampilkan plot langsung di notebook.

```python
# Konfigurasi output Bokeh untuk Jupyter
output_notebook()

print("âœ“ Bokeh siap untuk menampilkan plot di Jupyter notebook")
```

## 4. Create Basic Bokeh Plots

Membuat berbagai jenis plot Bokeh yang dapat digunakan di Jupyter atau diekspor ke Flask.

```python
# Plot 1: Time Series - Units Sold
p1 = figure(
    title="Units Sold Over Time",
    x_axis_label='Date',
    y_axis_label='Units Sold',
    x_axis_type='datetime',
    width=800,
    height=400,
    toolbar_location='right',
    tools='pan,wheel_zoom,box_zoom,reset,save'
)

p1.line(df['Date'], df['Units Sold'], line_width=2, color='steelblue', legend_label='Units Sold')
p1.circle(df['Date'], df['Units Sold'], size=4, color='steelblue', alpha=0.5)

# Add hover tool
hover1 = HoverTool(tooltips=[('Date', '@x{%F}'), ('Units', '@y{0,0}')], 
                   formatters={'@x': 'datetime'})
p1.add_tools(hover1)

print("âœ“ Plot 1 (Time Series) dibuat")
```

```python
# Plot 2: Scatter Plot - Price vs Units Sold
p2 = figure(
    title="Price vs Units Sold",
    x_axis_label='Price',
    y_axis_label='Units Sold',
    width=400,
    height=400,
    tools='pan,wheel_zoom,box_zoom,reset,save'
)

# Create ColumnDataSource untuk interaktivitas lebih baik
source = ColumnDataSource(data=dict(
    price=df['Price'],
    units=df['Units Sold'],
    category=df['Category'],
    date=df['Date'].astype(str)
))

scatter = p2.scatter('price', 'units', source=source, size=8, color='coral', alpha=0.6)

# Add hover tool
hover2 = HoverTool(tooltips=[
    ('Price', '@price{$0.00}'),
    ('Units', '@units{0,0}'),
    ('Category', '@category'),
    ('Date', '@date')
])
p2.add_tools(hover2)

print("âœ“ Plot 2 (Scatter Plot) dibuat")
```

```python
# Plot 3: Bar Chart - Category Performance
category_stats = df.groupby('Category')['Units Sold'].mean().sort_values(ascending=False)

p3 = figure(
    x_range=list(category_stats.index),
    title="Average Units Sold by Category",
    x_axis_label='Category',
    y_axis_label='Average Units Sold',
    width=400,
    height=400,
    toolbar_location='right',
    tools='pan,wheel_zoom,reset,save'
)

p3.vbar(x=list(category_stats.index), top=category_stats.values, width=0.6, color='green', alpha=0.7)

# Add hover tool
hover3 = HoverTool(tooltips=[('Category', '@x'), ('Avg Units', '@top{0.00}')])
p3.add_tools(hover3)

print("âœ“ Plot 3 (Bar Chart) dibuat")
```

## 5. Display Plots in Jupyter

Tampilkan plot secara langsung di notebook untuk preview.

```python
# Tampilkan plot 1
show(p1)
```

```python
# Tampilkan plot 2 dan 3 secara bersamaan (side-by-side)
from bokeh.layouts import row
show(row(p2, p3))
```

## 6. Export Plots untuk Flask (Components Method)

Metode ini memisahkan script Bokeh dari div-nya, ideal untuk embedding di template Flask.

```python
# Method 1: Menggunakan components() - Best Practice untuk Flask
# Ini memisahkan script Bokeh (yang hanya dimuat 1x) dari div (yang bisa di-render berkali-kali)

script, (div1, div2, div3) = components((p1, p2, p3))

print("=" * 60)
print("ğŸ“Š EXPORT UNTUK FLASK - COMPONENTS METHOD")
print("=" * 60)
print("\nPanjang Script Bokeh:", len(script), "karakter")
print("Panjang Div 1:", len(div1), "karakter")
print("Panjang Div 2:", len(div2), "karakter")
print("Panjang Div 3:", len(div3), "karakter")

print("\nâœ“ Siap untuk digunakan di Flask template")
print("\nCara penggunaan di Flask:")
print("1. Di app.py route:")
print("   script, (div1, div2, div3) = components((p1, p2, p3))")
print("   return render_template('my_dashboard.html', script=script, div1=div1, div2=div2, div3=div3)")
print("\n2. Di template my_dashboard.html:")
print("   {{ script | safe }}")
print("   {{ div1 | safe }}")
print("   {{ div2 | safe }}")
print("   {{ div3 | safe }}")
```

## 7. Preview Script dan Div HTML

Mari lihat struktur HTML yang akan digunakan di Flask template.

```python
# Preview structure dari script
print("=" * 60)
print("SCRIPT BOKEH (Script yang di-load di <head> atau akhir <body>)")
print("=" * 60)
print(script[:300] + "...\n")

print("=" * 60)
print("DIV HTML (Placeholder untuk plot)")
print("=" * 60)
print(div1[:200] + "...\n")

print("\nğŸ’¡ Tips:")
print("- Script Bokeh hanya perlu dimuat SEKALI per halaman")
print("- Bisa ada banyak div di halaman yang sama")
print("- Gunakan {{ script | safe }} di head atau akhir body")
print("- Gunakan {{ div | safe }} di content area")
```

## 10. Advanced: Tambahan Features

Tips dan tricks untuk meningkatkan interaktivitas.

```python
# Advanced Feature 1: Linking between plots
print("=" * 60)
print("ğŸ”— FEATURE 1: Linked Plots (Cross-filtering)")
print("=" * 60)

from bokeh.models import ColumnDataSource
from bokeh.layouts import row

# Create shared data source
shared_source = ColumnDataSource(df[['Date', 'Units Sold', 'Price', 'Category']])

# Plot dengan shared source
p_linked1 = figure(title="Units Over Time", x_axis_type='datetime', width=400, height=400)
p_linked1.line('Date', 'Units Sold', source=shared_source, line_width=2, color='steelblue')

p_linked2 = figure(title="Price vs Units", width=400, height=400)
p_linked2.scatter('Price', 'Units Sold', source=shared_source, size=8, color='coral')

print("âœ“ Plot yang 'linked' memungkinkan cross-filtering")
print("  - Ketika hover di satu plot, highlight di plot lain")
print("  - Shared ColumnDataSource membuat ini possible")

show(row(p_linked1, p_linked2))
```
